```python
from sympy import symbols, Eq, solve

# Define the symbols
x, y, z, w = symbols('x y z w', integer=True)

# Define the original equation
original_equation = Eq(59*x + 199*y + 287*z + 344*(100 - x - y - z), 10000)

# Substitute w
substituted_equation = original_equation.subs(w, 100 - x - y - z)

# Simplify the equation
simplified_equation = substituted_equation.simplify()

# Solve for z in terms of x and y
z_solution = solve(simplified_equation, z)[0]

# Define the constraints
max_x = 10000 // 59
max_y = 10000 // 199
max_z = 10000 // 287

# List to store valid solutions
valid_solutions = []
# Solve the equation with the constraints
for x_val in range(max_x + 1):
    for y_val in range(max_y + 1):
        if x_val + y_val >= 99: 
            continue
        z_val = z_solution.subs({x: x_val, y: y_val})
        if z_val.is_integer:
            if z_val + y_val + x_val >= 100: 
                continue
            z_val = int(z_val)
            w_val = 100 - x_val - y_val - z_val
            if 0 < z_val < max_z and 0 < w_val < 100:
                valid_solutions.append((x_val, y_val, z_val, w_val))

assert len(valid_solutions) == 1
assert valid_solutions[0][0] == 78
assert valid_solutions[0][1] == 13
assert valid_solutions[0][2] == 5
assert valid_solutions[0][3] == 4

print(f"Stamps:")
for i, d in enumerate([0.59, 1.99, 2.87, 3.44]):
      print(f"{valid_solutions[0][i]} @ ${d} ", end=' ')
```

```python
from sympy import symbols, Eq, solve, primerange

# Define the symbols
q, d, n, p = symbols('q d n p', integer=True)

# Define the equations based on the problem statement
eq1 = Eq(p, d * (d + n))
eq2 = Eq(p, q + 24)

# List of prime numbers (we'll consider primes up to a reasonable range)
primes = list(primerange(2, 100))

# Find the solution
valid_solutions = []
for q_val in primes:
    for d_val in primes:
        if d_val == q_val:
            continue
        for n_val in primes:
            if n_val == q_val or n_val == d_val:
                continue
            solution = solve((eq1.subs({d: d_val, n: n_val}), eq2.subs({q: q_val})), (p))
            if solution:
                p_val = solution[p]
                if p_val == d_val * (d_val + n_val):
                    valid_solutions.append((q_val, d_val, n_val, p_val))

valid_solutions
```

```python
from sympy import symbols, Eq, solve, primerange

# Define the symbols
p, q, d, n = symbols('p q d n', integer=True)
eq1 = Eq(p, d * (d + n))

# Simplified equation
simplified_eq = eq1.subs(p, q+24)

# List of prime numbers (we'll consider primes up to a reasonable range)
primes = list(primerange(2, 100))

# Find the solution
valid_solutions = []
for d_val in primes:
    for n_val in primes:
        if n_val == d_val:
            continue
        q_val = solve(simplified_eq.subs({d: d_val, n: n_val}), q)
        if q_val and q_val[0] in primes and q_val[0] != d_val and q_val[0] != n_val:
            valid_solutions.append((q_val[0], d_val, n_val, q_val[0] + 24))

valid_solutions
```

```python
from sympy import symbols, Eq, solve, S
from sympy.ntheory import isprime, primerange

# Define the symbols
q, d, n = symbols('q d n', integer=True)

# Simplified equation
simplified_eq = Eq(q, d * (d + n) - 24)

# Generate all prime number combinations where one prime is even and the other is odd
valid_solutions = []

# Use the prime range generator to create prime values
primes = list(primerange(2, 100))

# Algebraic filtering without explicit loops
for d_val in primes:
    for n_val in primes:
        if d_val != n_val and (d_val + n_val) % 2 != 0:
            q_val = simplified_eq.subs({d: d_val, n: n_val}).rhs
            if isprime(q_val):
                valid_solutions.append((q_val, d_val, n_val, q_val + 24))

valid_solutions
```
